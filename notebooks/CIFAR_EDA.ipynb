# CIFAR-10 Exploratory Data Analysis

This notebook is a thin orchestrator. All reusable logic lives in the `cifar10_eda` package under `src/`.

```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
%matplotlib inline

from cifar10_eda import (
    ARCHIVE_PATH,
    DATA_DIR,
    NUM_CLASSES,
    NUM_TEST,
    NUM_TRAIN,
    RANDOM_SEED,
    display_color_hists,
    drop_color_channel,
    extract_archive,
    load_cifar10,
    plot_class_distribution,
    set_seed,
    setup_logging,
    show_unique_images,
)
```

```python
# Configure logging and reproducibility
logger = setup_logging()
set_seed(RANDOM_SEED)
```

```python
# Extract the archive once (idempotent) and load the dataset
extracted_dir = extract_archive(ARCHIVE_PATH, DATA_DIR)
dataset = load_cifar10(extracted_dir)
```

```python
# Sanity report
import numpy as np

assert dataset.X_train.shape == (NUM_TRAIN, 3072), dataset.X_train.shape
assert dataset.y_train.shape == (NUM_TRAIN,), dataset.y_train.shape
assert dataset.X_test.shape == (NUM_TEST, 3072), dataset.X_test.shape
assert dataset.y_test.shape == (NUM_TEST,), dataset.y_test.shape
assert len(dataset.class_names) == NUM_CLASSES
assert dataset.X_train.dtype == np.uint8
assert dataset.X_test.dtype == np.uint8

logger.info("Train shape: %s, Test shape: %s", dataset.X_train.shape, dataset.X_test.shape)
logger.info("Class names: %s", dataset.class_names)
logger.info("Train label range: [%d, %d]", dataset.y_train.min(), dataset.y_train.max())
logger.info("Test label range: [%d, %d]", dataset.y_test.min(), dataset.y_test.max())
```

## Sample Images

Display the first occurrence of every class in the training set.

```python
unique_indices, _ = show_unique_images(
    dataset.X_train,
    dataset.y_train,
    class_names=dataset.class_names,
)
```

## Color-Channel Ablation

See what each class looks like when the red, green, or blue channel is zeroed out.

```python
drop_color_channel(
    dataset.X_train,
    dataset.y_train,
    indices=unique_indices,
    drop_color=0,  # red
    class_names=dataset.class_names,
);
```

```python
drop_color_channel(
    dataset.X_train,
    dataset.y_train,
    indices=unique_indices,
    drop_color=1,  # green
    class_names=dataset.class_names,
);
```

```python
drop_color_channel(
    dataset.X_train,
    dataset.y_train,
    indices=unique_indices,
    drop_color=2,  # blue
    class_names=dataset.class_names,
);
```

## Per-Channel Histograms

Distribution of pixel intensities for each RGB channel, one sample per class.

```python
display_color_hists(
    dataset.X_train,
    dataset.y_train,
    indices=unique_indices,
    class_names=dataset.class_names,
);
```

## Class Balance

```python
plot_class_distribution(
    dataset.y_train,
    dataset.y_test,
    class_names=dataset.class_names,
);
```
